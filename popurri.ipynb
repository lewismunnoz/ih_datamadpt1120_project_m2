```python
described_clarity = {
 
    'included': ['I1'],
    'internally_flawless': ['IF'],
    'very_slightly_included': ['VS2', 'VS1'],
    'very_very_slightly_included': ['VVS1', 'VVS2'],
    'slightly_included': ['SI1', 'SI2']
}

described_color = {
    "colorless": ["D", "E", "F"],
    "near_colorless": ["G", "H"],
    "slightly_tinted": ["I", "J"]
}
```

```python
diamonds['described_clarity'] = [key for clarity in diamonds.clarity for key, item in described_clarity.items() if clarity in item]
diamonds['described_color'] = [key for color in diamonds.color for key, item in described_color.items() if color in item]
```
